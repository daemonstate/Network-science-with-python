```python
#imports
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def read_adjacency_matrix(networkA):
    with open(networkA, 'r') as file:
        dimensions = file.readline().strip().split(',')
        num_vertices = int(dimensions[0])
        adjacency_matrix = np.zeros((num_vertices, num_vertices))
        for line in file:
            source, target, edgeweight = map(float, line.strip().split(','))
            source = int(round(source))
            target = int(round(target))
            adjacency_matrix[source, target] = edgeweight
    return adjacency_matrix
```

```python
file_paths = ['networkA.adj', 
              'networkB.adj', 
              'networkC.adj']
```

```python
def in_deg(adjacency_matrix):
    return np.sum(adjacency_matrix, axis=0)
def out_deg(adjacency_matrix):
    return np.sum(adjacency_matrix, axis=1)
def page_r(adjacency_matrix):
    G = nx.DiGraph(adjacency_matrix)
    pagerank = nx.pagerank(G)
    return list(pagerank.values())
def norm_deg(adjacency_matrix):
    G = nx.DiGraph(adjacency_matrix)
    degree_centrality = nx.degree_centrality(G)
    return list(degree_centrality.values())
def centr(adjacency_matrix):
    G = nx.DiGraph(adjacency_matrix)
    centralization = nx.algorithms.centrality.degree_alg.degree_centrality(G)
    return centralization
def get_top_vertices(values, N):
    top_indices = np.argsort(values)[-N:][::-1]
    top_values = [values[i] for i in top_indices]
    return list(zip(top_indices, top_values))
```

```python
for file_path in file_paths:
    if __name__ == "__main__":
        adj_matrix = read_adjacency_matrix(file_path)
        in_degrees = in_deg(adj_matrix)
        out_degrees = out_deg(adj_matrix)
        pagerank = page_r(adj_matrix)
        normalized_degree_centrality = norm_deg(adj_matrix)
        centralization = centr(adj_matrix)

        plt.figure(figsize=(16, 12))
        plt.subplot(2, 2, 1)
        plt.hist(in_degrees, bins=20, color='seagreen', edgecolor='black')
        plt.title("In-Degree Histogram")
        plt.xlabel("In-Degree")
        plt.ylabel("Frequency")
        
        plt.subplot(2, 2, 2)
        plt.hist(out_degrees, bins=20, color='ivory', edgecolor='black')
        plt.title("Out-Degree Histogram")
        plt.xlabel("Out-Degree")
        plt.ylabel("Frequency")
        
        plt.subplot(2, 2, 3)
        plt.hist(pagerank, bins=20, color='orangered', edgecolor='black')
        plt.title("PageRank Histogram")
        plt.xlabel("PageRank")
        plt.ylabel("Frequency")
        
        plt.subplot(2, 2, 4)
        plt.hist(normalized_degree_centrality, bins=20, color='teal', edgecolor='black')
        plt.title("Normalized Degree Centrality Histogram")
        plt.xlabel("Normalized Degree Centrality")
        plt.ylabel("Frequency")
        plt.tight_layout()
        plt.show()

        print("Centralization: ", centralization)

        top_10_in_degree = get_top_vertices(in_degrees, 10)
        top_10_out_degree = get_top_vertices(out_degrees, 10)
        top_10_pagerank = get_top_vertices(pagerank, 10)
        top_10_normalized_degree_centrality = get_top_vertices(normalized_degree_centrality, 10)

        print("10 vertices with the highest in-degree:")
        for vertex, in_degree_value in top_10_in_degree:
            print(f"Vertex {vertex}: In-degree = {in_degree_value}")

        print("10 vertices with the highest out-degree:")
        for vertex, out_degree_value in top_10_out_degree:
            print(f"Vertex {vertex}: Out-degree = {out_degree_value}")

        print("10 vertices with the highest PageRank:")
        for vertex, pagerank_value in top_10_pagerank:
            print(f"Vertex {vertex}: PageRank = {pagerank_value}")

        print("10 vertices with the highest normalized degree centrality:")
        for vertex, degree_centrality_value in top_10_normalized_degree_centrality:
            print(f"Vertex {vertex}: Normalized Degree Centrality = {degree_centrality_value}")
```
